```python
def cosine_similarity_matrix(matrix):
    similarity_matrix = cosine_similarity(matrix)
    return pd.DataFrame(similarity_matrix, index=matrix.index, columns=matrix.index)

def recommender(user, active_data, inactive_data, n_items=10, neighbor_size=3):
    # 현재 사용자의 아이템별 소비패턴을 활성고객 데이터로부터 가져오기
    user_pattern = active_data.loc[user]
    
    # 활성고객들과의 유사도 행렬 생성
    similarity_matrix = cosine_similarity_matrix(active_data)
    
    # 현재 사용자와 유사한 상위 neighbor_size명의 활성고객 선택
    similar_users = similarity_matrix[user].sort_values(ascending=False)[1:neighbor_size+1].index
    
    # 유사한 활성고객들의 아이템별 평균 소비패턴 계산
    similar_users_pattern = active_data.loc[similar_users].mean()
    
    # 추천할 아이템 선택
    unrated_items = inactive_data.loc[user][inactive_data.loc[user] == 0].index
    recommended_items = (user_pattern + similar_users_pattern) / 2
    recommended_items = recommended_items[unrated_items].sort_values(ascending=False)[:n_items]
    
    return recommended_items

# 카드 추천 함수 부르기
user_id = "FR1nNGOO"  # 예시로 회원번호 "1003"을 대상으로 추천 수행
recommended_cards = recommender(user_id, active_customers, inactive_customers)
if isinstance(recommended_cards, str):
    print(recommended_cards)
else:
    print("추천하는 카드:")
    print(recommended_cards)
```
